# Coulomb Matrix Representation

This notebook builds the Coulomb matrix molecular representation
for use in the Δ-ML model.

## What is the Coulomb Matrix?

For a molecule with N atoms, the Coulomb matrix M is an N×N matrix:

- **Diagonal:** M_ii = 0.5 · Z_i^2.4 (nuclear charge of atom i)
- **Off-diagonal:** M_ij = Z_i · Z_j / |R_i - R_j| (electrostatic interaction)

Where Z_i is the atomic number and R_i is the 3D position of atom i.

## Why do we need it?

ML models need fixed-size numerical inputs. Raw molecular structures
(atom types + 3D coordinates) have variable size — different molecules
have different numbers of atoms. The Coulomb matrix encodes the full
3D structure as a matrix, which we then zero-pad and sort to get
a fixed-size representation.
